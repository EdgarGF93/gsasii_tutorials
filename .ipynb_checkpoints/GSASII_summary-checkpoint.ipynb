#### Importing the module using the shortcut

```python
%matplotlib inline
import G2script as G2
from GSASIIplot import PlotPatterns, PlotPowderLines
import matplotlib.pyplot as plt
import numpy as np
```

#### Create new project (.gpx)

```python
PROJECT = 'projects/gsas_script.gpx'
gpx = G2.G2Project(newgpx=PROJECT)
gpx.save()
```

#### We will create an histogram for three CIF files and a whole data containing the diffraction peaks from the three phases

##### Add Phases from .cif files

```python
CIF_ZnO = 'cif/ZnO.cif'
CIF_Al2O3 = 'cif/Al2O3.cif'
CIF_CaF2 = 'cif/CaF2.cif'
```

```python
phase_zno = gpx.add_phase(
    phasefile=CIF_ZnO,
    phasename=CIF_ZnO,
)

phase_al2o3 = gpx.add_phase(
    phasefile=CIF_Al2O3,
    phasename=CIF_Al2O3,
)

phase_caf2 = gpx.add_phase(
    phasefile=CIF_CaF2,
    phasename=CIF_CaF2,
)
```

```python
print("*** Phases ***")
for ph in gpx.phases():
    print(f'Phase : {ph.name}, composition: {ph.composition}, density: {ph.density:.2f}')
```

```python
for atom in phase_al2o3.atoms():
    print(f'The atom {atom.element} has cx {atom.cx}, cs {atom.cs}, ct {atom.ct}, cia {atom.cia}, coordinates {atom.coordinates}')
```

#### Simulate Powder Diffraction with the CIF phases
To simulate or import diffraction data, we have to include a .instprm file with some instrumental parameters (wavelength, polarization...)

```python
hist_zno = gpx.add_simulated_powder_histogram(
    histname='hist_zno', iparams='cuka.instprm', Tmin=20, Tmax=100,
    phases=[
        phase_zno,
    ],
)
hist_caf2 = gpx.add_simulated_powder_histogram(
    histname='hist_caf2', iparams='cuka.instprm', Tmin=20, Tmax=100,
    phases=[
        phase_caf2,
    ],
)
hist_al2o3 = gpx.add_simulated_powder_histogram(
    histname='hist_al2o3', iparams='cuka.instprm',Tmin=20,Tmax=100,
    phases=[
        phase_al2o3,
    ],
)
hist_all = gpx.add_simulated_powder_histogram(
    histname='hist_all', iparams='cuka.instprm', Tmin=20, Tmax=100,
    phases=[
        phase_al2o3,
        phase_zno,
        phase_caf2,
    ],
)

gpx.do_refinements(
    refinements=[{}],
    histogram='all',
    phase='all',
)
```

#### We have simulated the three isolated phases and a mixture of the three

```python
%matplotlib widget
fig, axes = plt.subplots(ncols=4, figsize=(12,4))
HISTS = [hist_zno, hist_al2o3, hist_caf2, hist_all]
LABELS = ['hist_zno', 'hist_al2o3', 'hist_caf2', 'hist_all']
XLIMS = [10,120]
for ind in range(4):
    hist = HISTS[ind]
    axes[ind].plot(hist.getdata("X"), hist.getdata("ycalc"), label=LABELS[ind])
    axes[ind].legend()
    axes[ind].set_xlim(XLIMS)
```

#### Add Powder Diffraction Data

#### Create an histogram with PD data and linked to the imported phases

```python
pwd = gpx.add_powder_histogram(
    datafile='diff_data/cpd-1h.prn', 
    iparams='cuka.instprm',
    phases=gpx.phases(),
)
```

#### Perform a simple refinement

```python
gpx.do_refinements()
```

```python
fig, axes = plt.subplots(ncols=4, nrows=1, figsize=(12,3))
data = pwd.data['data'][1]
LABELS = ['Obs', 'Calc', 'Background', 'Residual']
DATA_IND = [1,3,4,5]
for num in range(4):
    axes[num].plot(data[0], data[DATA_IND[num]], label=LABELS[num])
    axes[num].legend()
plt.show()
```

#### Refine the background

```python
background_refinement = {
    "Background" : {
        "no. coeffs" : 10,
        "type" : "chebyschev-1",
        "refine" : True,
    }
}
pwd.set_refinements(background_refinement)
```

```python
gpx.do_refinements()
```

```python
%matplotlib widget
fig, ax = plt.subplots(figsize=(10,4))
ax.plot(pwd.getdata('X'), pwd.getdata('Yobs'), label="Data")
ax.plot(pwd.getdata('X'), pwd.getdata('Ycalc'), label="Calculated")
ax.set_xlabel('2theta(deg)')
ax.set_ylabel('Intensity')
ax.legend()
pass
```

#### Refine sequentially the rest of parameters

```python
ref_list = [
    {"set" : {'Limits' : [0,100]}},
    {"set" : {"LeBail" : True}},
    {"set" : {'Cell' : True, "LeBail" : False}},
    {"set" : {'Instrument Parameters' : ['U', 'V', 'W', 'X', 'Y', 'Z']}},
    {"set" : {"Scale" : True, "Use" : True}},
]
gpx.do_refinements(ref_list)
```
